```python
def standardize1(v):
  eqn = (v - v.mean())/v.std()
  return eqn
```

```python
sns.scatterplot(x = standardize1(mod1.fittedvalues),y = standardize1(mod1.resid) )
plt.xlabel("Standardized fitted values")
plt.ylabel("standardized residual values")
```

```python
import re 
import string

def clean_text(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub("[0-9" "]+"," ",text)
    text = re.sub('[]', '', text)
    return text  

clean = lambda x: clean_text(x)

```

```python
data['workclass'] = data.workclass.apply(clean)
data.workclass
```

```python
def split_into_words(i):
    return (i.split(" "))
```

```python
from sklearn.model_selection import train_test_split

email_train,email_test = train_test_split(data,test_size=0.3)
```

```python
 
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
vectorizer = CountVectorizer(min_df = 1, max_df = 5)
X = vectorizer.fit_transform(data, y=None)
word_freq_df = pd.DataFrame({'term': vectorizer.get_feature_names_out(), 'occurrences':np.asarray(X.sum(axis=0)).ravel().tolist()})
word_freq_df['frequency'] = word_freq_df['occurrences']/np.sum(word_freq_df['occurrences'])
#print(word_freq_df.sort('occurrences',ascending = False).head())
```

```python
emails_bow = CountVectorizer(analyzer=split_into_words).fit(data.education)
```

```python
all_emails_matrix = emails_bow.transform(data.education)
all_emails_matrix.shape
```

```python
all_emails_matrix.toarray()
```

```python
train_emails_matrix = emails_bow.transform(email_train.education)
train_emails_matrix.shape

# For testing messages
test_emails_matrix = emails_bow.transform(email_test.education)
test_emails_matrix.shape 
```

```python
from sklearn.naive_bayes import MultinomialNB as MB


# Multinomial Naive Bayes
classifier_mb = MB()
classifier_mb.fit(train_emails_matrix,email_train.Salary)

train_pred_m = classifier_mb.predict(train_emails_matrix)
accuracy_train_m = np.mean(train_pred_m==email_train.Salary) # 98%

test_pred_m = classifier_mb.predict(test_emails_matrix)
accuracy_test_m = np.mean(test_pred_m==email_test.Salary)
```
